```python
import numpy as np
import matplotlib.pyplot as plt

def create_arr(length):
    df = np.zeros(length)
    for x in range (df.size):
        df[x] = df[x-1] + np.random.randint(-1, 2)
    return df
def create_graph(arr):
    x = range(0, arr.size)
    y = arr
    plt.title("Plot")
    plt.xlabel("X axis")
    plt.ylabel("Y axis")
    plt.plot(x, y, color ="blue")
    plt.show()
def min_max(arr):
    return f"Min: {arr.min()} Max: {arr.max()}"
```

```python
df = create_arr(100)
print(df)
```

```python
create_graph(df)
```

```python
print(min_max(df))
```

```python
for x in range (50):
    temp = create_arr(100)
    if temp.max() > 20:
        print (f"nr {x} exc. 20")
        create_graph(temp)
```
