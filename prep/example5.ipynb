```python
import numpy as np
import matplotlib.pyplot as plt

def create_array(length):
    df = np.zeros(length)
    for x in range (df.size):
        df[x] = df[x-1] + np.random.randint(-1,2)
    return df
def create_graph(g_df):      
    x = np.arange(0, g_df.size)
    y = g_df
    plt.title("Plot")
    plt.xlabel("X axis")
    plt.ylabel("Y axis")
    plt.plot(x, y, color ="blue")
    plt.show()
def max_min(m_df):
    print(f"Max: {m_df.max()} Min: {m_df.min()}")

df = create_array(100)
create_graph(df)
max_min(df)
for x in range(50):
    df_temp = create_array(100)
    if df_temp.max() > 20:
        print(f"nr {x} simulation has an array exceeding 20")
        create_graph(df_temp)
```
