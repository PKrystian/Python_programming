# 4. Zadanie

```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame

data1 = {'1': [4, None, 8, 8, 12],
        '2': [2, 4, 6, 45, None],
        '3': [2, None, 8, 6, 70],
        '4': [2, 4, 6, None, 40],
        '5': [None, 4, 6, 8, 10]}
frame1 = pd.DataFrame(data1)
print('Before changes: ', '\n', frame1, '\n')
for i in range(5):
    xd = (frame1.iloc[:,[i]].sum())/len(frame1.iloc[:,[i]])
    frame1.iloc[:,[i]] = frame1.iloc[:,[i]].fillna(xd)
print('After changes: ' , '\n', frame1)
```
