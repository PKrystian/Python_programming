# 5. Zadanie

```python
import numpy as np

arr1 = np.array([[46.21, 18.21], [29.21, 21.21]])
arr2 = np.array([[11.21, 19.21], [54.63, 12.83]])
i, j, sum_arr1, sum_arr2, pro_arr1, pro_arr2 = 0, 0, 0, 0, 1, 1
for j in range(len(arr1)):
    for i in range(len(arr1)):
        sum_arr1 = sum_arr1 + arr1[i][j]
        pro_arr1 = pro_arr1 * arr1[i][j]
for j in range(len(arr2)):
    for i in range(len(arr2)):
        sum_arr2 = sum_arr2 + arr2[i][j]
        pro_arr2 = pro_arr2 * arr2[i][j]
print ('Sum: ', sum_arr1 + sum_arr2)
print ('Product: ', pro_arr1 * pro_arr2)
```
