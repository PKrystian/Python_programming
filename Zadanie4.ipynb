```python
class Rectangle:
    side_a = int(input("Podaj podstawe : "))
    side_h = int(input("Podaj wysokosc : "))
    area_rectangle = (side_a * side_h) / 2
    print("Pole trojkata to: ",area_rectangle,"")
class Square(Rectangle):
    area_square = Rectangle.side_a * Rectangle.side_a
    print("Pole kwadratu to: ",area_square,"")
```
